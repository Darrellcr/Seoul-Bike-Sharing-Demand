## Elastic Net

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder, FunctionTransformer, PowerTransformer, StandardScaler, QuantileTransformer
from sklearn.linear_model import ElasticNetCV, ElasticNet
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
import warnings
warnings.filterwarnings('default')
```

```python
target = 'Rented Bike Count'
numerical_feats = rented_bike.loc[:, rented_bike.columns != target].select_dtypes(include=np.number).columns
# numerical_feats = ['Rainfall', 'Snowfall', 'Visibility', 'Solar Radiation']
fig, axs = plt.subplots(3, 3, figsize=(8, 6))
for ax, feat in zip(axs.flat, numerical_feats):
    ax.hist(rented_bike[feat])
    ax.set_title(feat)

plt.tight_layout()
plt.show()
```

```python
x = (rented_bike
         .query("`Functioning Day`")
         .loc[:, ~rented_bike.columns.isin(['Rented Bike Count', 'DateTime', 'Functioning Day', 'Snowfall', 'Rainfall', 'Dew point temperature'])])
y = rented_bike.query("`Functioning Day`")['Rented Bike Count']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=7)

categorical_features = ['Seasons']
numerical_features = ['Hour', 'Temperature', 'Humidity', 'Wind speed']
transformed_features = ['Solar Radiation', 'Visibility']
transformers = [
    ('transformer', Pipeline([
        ('logtransformer', FunctionTransformer(np.log1p)),
        ('normalizer', MinMaxScaler())
    ]), transformed_features),
    ('scaler', MinMaxScaler(), numerical_features),
    ('encoder', OneHotEncoder(), categorical_features),
]
preprocessor = ColumnTransformer(
    transformers=transformers,
    remainder='passthrough'
)

ratios = np.arange(0.25, 1, 0.005)
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', ElasticNetCV(l1_ratio=ratios, 
                           alphas=None, 
                           cv=10, 
                           random_state=7, 
                           n_jobs=4,
                           tol=1e-10
                          ))
])

res = pipeline.fit(x_train, y_train)
y_pred = pipeline.predict(x_test)

print(f'{r2_score(y_test, y_pred) = }')
print(f'{mean_squared_error(y_test, y_pred) = }')
print(f'{mean_absolute_error(y_test, y_pred) = }')
```

jika melihat metrics yang didapat setelah menggunakan elastic net, dapat dilihat bahwa penggunaannya kurang cocok dengan data ini

```python
print('alpha=' + str(pipeline.named_steps['model'].alpha_))
print('l1 ratio=' + str(pipeline.named_steps['model'].l1_ratio_))
print('intercept=' + str(pipeline.named_steps['model'].intercept_))
print('coef=' + str(pipeline.named_steps['model'].coef_))
```

```python
features = list(preprocessor.feature_names_in_)
season_index = features.index('Seasons')
seasons = preprocessor.named_transformers_['encoder'].categories_[0]
features.pop(season_index)
for i, season in enumerate(seasons):
    features.insert(season_index + i, f'Season_{season}')
feature_coefs = list(zip(
    features,
    pipeline.named_steps['model'].coef_
))
feature_coefs
```

```python
sorted_coefs = sorted(feature_coefs, key=lambda f: abs(f[1]))
unzipped = list(zip(*sorted_coefs))

fig, ax = plt.subplots()
h_bar = ax.barh(unzipped[0], np.abs(unzipped[1]), left=10)
labels = [f'{x:.2f}' for x in unzipped[1]]
ax.bar_label(h_bar, labels=labels)
ax.set_title('Feature Coefficient')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):

    #Determing rolling statistics
    rolmean = timeseries.rolling(30).mean()
    rolstd = timeseries.rolling(30).std()

    #Plot rolling statistics:
    fig = plt.figure(figsize=(12, 8))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput) 
```

```python
test_stationarity(imputed)
```

```python
first_difference = imputed.diff()
test_stationarity(first_difference.dropna())
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, axs = plt.subplots(2, 1, figsize=(9,6))
lags = 30
plot_acf(first_difference.dropna(), lags=lags, zero=False, auto_ylims=True, ax=axs[0])
axs[0].set_title('Auto Correlation')
plot_pacf(first_difference.dropna(), lags=lags, zero=False, auto_ylims=True, ax=axs[1])
axs[1].set_title('Partial Auto Correlation')

fig.tight_layout()
plt.show()
```

## Rolling Forecast ARIMA

```python
order = ([*np.arange(1,7), 12, 28], 1, 0)
test_size = int(len(imputed) * .3)
end_train = imputed.shape[0] - test_size
train = imputed[:end_train].asfreq('D')
test = imputed[end_train:].asfreq('D')
history = [x for x in train]
predictions = []

for obs in test:
    model = ARIMA(history, order=order)
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    history.append(obs)

rmse = np.sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % rmse)

predictions = pd.Series(predictions, index=test.index)
```

```python
plt.figure(figsize=(13,4))
plt.plot(test)
plt.plot(predictions, color='red')
labels = [
    str(month) + '-' + str(day) for month, day in
    (zip(test.index.month, test.index.day))
]
plt.legend(['True', 'Prediction'])
plt.suptitle('Rolling Forecast Arima Test', y=.95)
plt.xticks(ticks=test.index[::6], labels=labels[::6])
plt.show()
```
